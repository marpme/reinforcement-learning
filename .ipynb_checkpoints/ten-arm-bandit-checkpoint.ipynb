```python
import numpy as np
import matplotlib.pyplot as plt

# set up all initial values
NUMBER_OF_BANDITS = 10
NUMBER_OF_STEPS = 1000
NUMBER_OF_RUNS = 5000
```

```python

def get_bandit_function(bandits_means, bandits_sigma=1):
    '''
    Creates a normal distrubuted bandits based on the given means
    
    bandits_means:
        determine the mean of the certain normal distributed bandit
    bandits_sigma = 1:
        determines the sigma for normal distrubtions
        
    get_action_rewards: 
        Function which gets you an array of 1000 steps for each action you have choosen
    '''
    rewardsArray = np.zeros((len(bandits_means), NUMBER_OF_STEPS))
    for index in range(len(bandits_means)):
        rewardsArray[index] = np.random.normal(bandits_means[index], bandits_sigma, NUMBER_OF_STEPS)
            
    def get_action_rewards(action):
        return rewardsArray[action]
    
    return get_action_rewards
```

```python
def get_average_rewards(epsilon=0.1, optimistic=0., nb_runs = 2000, nb_steps=1000):
    """
    Compute the average reward for different 10-armed bandit problems.
    
    Parameters
    ----------
    epsilon : float
        epsilon: probability of chosing a random action (exploration).
    optimistic: float
        Initial value of all action-values.
    nb_runs: int
        Number of runs with different 10-armed bandits for the averaging.
    nb_steps: int
        Number of steps. How often each bandit is played.
        
    Returns
    -------
    numpy array (dtype: double)
        averaged rewards. i-th element in the array contains the average reward at iteration step i.
    numpy array (dtype: double)
        averaged percentage of using the optimal action. i-th element in the array contains the average reward at iteration step i.    
    """
     
    avgRewards = np.zeros(nb_steps)
    bestActions = np.zeros(nb_steps)
    
    usedActions = np.zeros(nb_steps, dtype=int)
    bandits_rewards = np.zeros((10,nb_steps))
            
    for run in range(nb_runs):
        banditsArray = np.random.normal(0, 1, NUMBER_OF_BANDITS)
        action_reward = get_bandit_function(banditsArray, 1)
        bestActionIndex = np.argmax(banditsArray)
        
        QAs = np.full(10, optimistic)
        NAs = np.zeros(10)
        
        for i in range(10):
            bandits_rewards[i] = action_reward(i)
        
        for time in range(nb_steps):
            
            if np.random.random() < epsilon:
                # exploration
                index = np.random.choice(NUMBER_OF_BANDITS)
                # print('[EXP] Best bandit ', bestActionIndex, ' choosen bandit ', index)
                reward = bandits_rewards[index][time]
                NAs[index] += 1
                QAs[index] = QAs[index] + (1/NAs[index]) * (reward - QAs[index])


                avgRewards[time] += bandits_rewards[index][time]
                bestActions[time] += 1 if bestActionIndex == index else 0
            else:
                # greedy
                index = np.argmax(QAs)
                # print('[GREED] Best bandit: ', bestActionIndex, ' choosen bandit: ', index, " QAB: ", QAs[bestActionIndex], " QAC: ", QAs[index] )
                reward = bandits_rewards[index][time]
                NAs[index] += 1
                QAs[index] = QAs[index] + (1/NAs[index]) * (reward - QAs[index])


                avgRewards[time] += bandits_rewards[index][time]
                bestActions[time] += 1 if bestActionIndex == index else 0
        
    # print(avgRewards)
    # print(bestActions)
    avgRewards /= nb_runs
    bestActions /= nb_runs
    return (avgRewards, bestActions)
```

```python
(avgRewards_e01, bestActions_e01) = get_average_rewards(0.01, 0., NUMBER_OF_RUNS, NUMBER_OF_STEPS)
print('run 1')
(avgRewards_e05, bestActions_e05) = get_average_rewards(0.05, 0., NUMBER_OF_RUNS, NUMBER_OF_STEPS)
print('run 2')
(avgRewards_e1, bestActions_e1) = get_average_rewards(0.1, 0., NUMBER_OF_RUNS, NUMBER_OF_STEPS)
print('run 3')
(avgRewards_e0, bestActions_e0) = get_average_rewards(0, 0., NUMBER_OF_RUNS, NUMBER_OF_STEPS)
print('run 4')
```

```python
plt.plot(np.arange(NUMBER_OF_STEPS), avgRewards_e1, label='$\epsilon=0.1$')
plt.plot(np.arange(NUMBER_OF_STEPS), avgRewards_e05, label='$\epsilon=0.05$')
plt.plot(np.arange(NUMBER_OF_STEPS), avgRewards_e01, label='$\epsilon=0.01$')
plt.plot(np.arange(NUMBER_OF_STEPS), avgRewards_e0, label='greed')

plt.legend(loc=4)
plt.xlabel("steps")
plt.ylabel("average reward")
plt.xlim(1,NUMBER_OF_STEPS)
plt.ylim(0, 1.7)
plt.show()
```

```python
plt.plot(np.arange(NUMBER_OF_STEPS), bestActions_e1, label='$\epsilon=0.1$')
plt.plot(np.arange(NUMBER_OF_STEPS), bestActions_e05, label='$\epsilon=0.05$')
plt.plot(np.arange(NUMBER_OF_STEPS), bestActions_e01, label='$\epsilon=0.01$')
plt.plot(np.arange(NUMBER_OF_STEPS), bestActions_e0, label='greedy')

plt.legend(loc=4)
plt.ylim(0., 1.05)
plt.xlim(1,NUMBER_OF_STEPS)
plt.xlabel("steps")
plt.ylabel("% optimal action")
plt.show()
```
